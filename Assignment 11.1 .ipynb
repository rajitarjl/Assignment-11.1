```python
#Import 
import matplotlib.pyplot as plt
from urllib.request import urlretrieve
 

# Import pandas
import pandas as pd

# Assign url of file: url
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'

urlretrieve(url, 'titanic.csv')

df= pd.read_csv('titanic.csv', sep=',')
```

```python
df.head()
```

```python

# Data to plot
counts = df['sex'].value_counts()
counts
```

```python
labels = 'Male', 'Female'
labels
```

```python
sizes = df['sex'].value_counts()
colors = ['gold', 'lightskyblue']
explode = (0, 0)  # explode 1st slice
 
# Plot a piechart to show male / female proportions
plt.title("Pie Chart to Represent Male/Female Distribution")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(loc='best', bbox_to_anchor=(1, 0.5), shadow=True, ncol=1)
plt.axis('equal')
plt.show()
```

```python

# Plot a Scatter Plot with Fare paid vs Age,\n color by gender 
df_male = df[df.sex == 'male']
plt.scatter(df_male.age, df_male.fare,alpha=0.5, c="green",label="Male")
df_female = df[df.sex == 'female']
plt.scatter(df_female.age,df_female.fare,alpha=0.5, c="blue",label="Female")
plt.title("Scatter Plot with Fare paid vs Age,\n color by gender ")
plt.legend(loc=2, bbox_to_anchor=(1, 0.5), shadow=True, ncol=1)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
